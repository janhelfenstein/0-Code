```python
# numpy arrays

ar1 = np.linspace(0, 5, 51)
print(ar1)

ar2 = np.arange(0, 10, 1) # 10 will not be in the array
ar21 = np.arange(0, 10.01, 1) # 10 will be in the array
ar22 = np.arange(10)
ar23 = np.arange(0, 10, 2)
print(ar2, "and", ar21, "and", ar22, "and", ar23)

ar3 = np.array([1, 2, 3, 4]) 
ar4 = np.array([2, 3, 4, 5])
ar5 = ar3 * ar4
ar6 = ar3 * 2
print(ar3, "\n", ar4, "\n", ar5, "\n", ar6)
```

```python
# matrices
A = np.identity(3)
B = np.matrix(3)
C = np.matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
print(A, "\n", B, "\n", C, "\n", A+C)

print(A[1]) # print second row
print(A[0][2]) # print A_13, third item on first column

print("Trace of matrix A is", np.trace(A))
print("Determinant of matrix \n", A, "\n is \n", np.linalg.det(A)) # det = 1
print("Determinant of matrix \n", C, "\n is \n", np.linalg.det(C)) # det = 0
```

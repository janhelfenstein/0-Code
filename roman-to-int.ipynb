Given a roman numeral, convert it to an integer.
Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

Symbol       Value
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
For example, 2 is written as II in Roman numeral, just two ones added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

I can be placed before V (5) and X (10) to make 4 and 9. 
X can be placed before L (50) and C (100) to make 40 and 90. 
C can be placed before D (500) and M (1000) to make 400 and 900.

```python
def RomanToInt(s):
    """
    :type s: str
    :rtype: int
    """
    num = 0
    # create list from letters in string s
    ls = list(s)
    for i in range(len(ls)):
        if ls[i] == 'M':
            if i >= 1 and ls[i-1] == "C":
                num += 900
            else: num += 1000
        elif ls[i] == "C":
            continue
        else:
            break

    for i in range(len(ls)):
        if ls[i] == 'D':
            if i >= 1 and ls[i-1] == "C":
                num += 400
            else: num += 500
        elif ls[i] == "C":
            continue
        else:
            break


    for i in range(len(ls)):
        if ls[i] == 'C':
            if i >= 1 and ls[i-1] == "X":
                num += 90
            else: num += 100
        elif ls[i] == "X":
            continue
        else:
            break

    for i in range(len(ls)):
        if ls[i] == 'L':
            if i >= 1 and ls[i-1] == "X":
                num += 40
            else: num += 50
        elif ls[i] == "X":
            continue
        else:
            break    

    for i in range(len(ls)):
        if ls[i] == 'X':
            if i >= 1 and ls[i-1] == "I":
                num += 9
            else: num += 10
        elif ls[i] == "I":
            continue
        else:
            break

    for i in range(len(ls)):
        if ls[i] == 'V':
            if i >= 1 and ls[i-1] == "I":
                num += 4
            else: num += 5
        elif ls[i] == "I":
            continue
        else:
            break
        
    return num



"""
    for i in range(len(ls)):
        if ls[i] == 'M':
            if i >= 1 and ls[i-1] == "C":
                num += 900
            else: num += 1000
        elif ls[i] == 'D':
            if i >= 1 and ls[i-1] == "C":
                num += 400
            else: num += 500
        elif ls[i] == 'C':
            if i >= 1 and ls[i-1] == "X":
                num += 90
            else: num += 100
        if ls[i] == 'L':
            if i >= 1 and ls[i-1] == "X":
                num += 40
            else: num += 50
        elif ls[i] == 'X':
            if i >= 1 and ls[i-1] == "I":
                num += 9
            else: num += 10
        elif ls[i] == 'V':
            if i >= 1 and ls[i-1] == "I":
                num += 4
            else: num += 5
        elif ls[i] == "I":
                num += 1
    return num
"""

# test cases
s = "III"
ls = list(s)
print(ls)
length = len(ls)
print(length)
print(RomanToInt(s))

s = "LVIII"
print(RomanToInt(s))

s = "MCMXCIV"
print(RomanToInt(s))
```

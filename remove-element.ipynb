Remove element

Given an integer array nums and an integer val, remove all occurrences of val in nums in-place. The order of the elements may be changed. Then return the number of elements in nums which are not equal to val.

Consider the number of elements in nums which are not equal to val be k, to get accepted, you need to do the following things:

- Change the array nums such that the first k elements of nums contain the elements which are not equal to val. The remaining elements of nums are not important as well as the size of nums.
- Return k.

```python
def removeElement(nums, val):
    array = []
    k = len(nums)
    for i in range(0, len(nums)-1):
        if nums[i] == val:
            k -= 1
        else:
            array.append(nums[i])
    return k, array

# test cases
nums = [3,2,2,3]
val = 3
k1, array1 = removeElement(nums, val)
print(k1)
print(array1)

nums = [0,1,2,2,3,0,4,2]
val = 2
k2, array2 = removeElement(nums, val)
print(k2)
print(array2)
```

solution with append

```python
def removeElement(nums, val):
    """
    :type nums: List[int]
    :type val: int
    :rtype: int
    """        
    array = []
    k = 0
    for i in range(0, len(nums)):
        if nums[i] != val:
            k += 1
            array.append(nums[i])
    for i in range(len(array)):
        nums[i] = array[i]
    return k

# test cases
nums = [3,2,2,3]
val = 3
k1 = removeElement(nums, val)
print(k1)
print(nums)

nums = [0,1,2,2,3,0,4,2]
val = 2
k2 = removeElement(nums, val)
print(k2)
print(nums)

nums = [3,3]
val = 3
k3 = removeElement(nums, val)
print(k3)
print(nums)
```

method without creating a new array

```python
def removeElement(nums, val):
    """
    :type nums: List[int]
    :type val: int
    :rtype: int
    """        
    k = 0 # counter for number of elements in array that are not equal to val
    for i in range(0, len(nums)):
        if nums[i] != val:
            nums[k] = nums[i]
            k += 1
    return k

# test cases
nums = [3,2,2,3]
val = 3
k1 = removeElement(nums, val)
print(k1)
print(nums)

nums = [0,1,2,2,3,0,4,2]
val = 2
k2 = removeElement(nums, val)
print(k2)
print(nums)

nums = [3,3]
val = 3
k3 = removeElement(nums, val)
print(k3)
print(nums)
```

My solution to the Two Sum problem on 
https://leetcode.com/problems/two-sum/
Brute Force

```python

def twoSum(nums, target):
    results = []
    for i in range(0, len(nums)-1):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == target:
                results.append(i)
                results.append(j)
    return results

# test cases
nums1 = [2, 7, 11, 15]
target1 = 9
nums2 = [3, 2, 4]
target2 = 6
nums3 = [3, 3]
target6 = 6
a1 = twoSum(nums1, target1)
a2 = twoSum(nums2, target2)
a3 = twoSum(nums3, target6)
print(a1, a2, a3)
```

Solution to the problem, if we know that only 1 solution-pair exists within the list:

```python
def twoSum(nums, target):
    for i in range(0, len(nums)-1):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == target:
                    return [i, j]
    return [] # return empty list if no solution is found

# test cases
nums1 = [2, 7, 11, 15]
target1 = 9
nums2 = [3, 2, 4]
target2 = 6
nums3 = [3, 3]
target6 = 6
a1 = twoSum(nums1, target1)
a2 = twoSum(nums2, target2)
a3 = twoSum(nums3, target6)
nums4 = [2, 7, 11, 15, 11, 11, 11, 7, 2, 7, 15, 7]
target4 = 9
a4 = twoSum(nums4, target4)
print(a1, a2, a3, a4)
```

Hashmap solution to the TwoSum problem
target = 9
logic:
iterate through list by indices (from 0 to len(nums))
x is your current number in the list.
define y as
y = target - x


```python
def twoSum(nums, target):
    h = {} # define hashmap - value : index

    # first, create the hashmap. Store the value and the locations of the values together
    for i in range(len(nums)):
        h[nums[i]] = i
    # for example, for the value 3, store all the positions in nums that have value 3.

    # now, go through the list and find partners
    for i in range(0, len(nums)):
        x = nums[i]
        y = target - x
        # check if y exists in the hashmap. Skip the location of x.
        if y in h and h[y] != i:
            return(i, h[y])
    # return empty list if no valid pair is found
    return []


# test cases
nums1 = [2, 7, 11, 15]
target1 = 9
nums2 = [3, 2, 4]
target2 = 6
nums3 = [3, 3]
target3 = 6
nums4 = [2, 7, 11, 15, 11, 11, 11, 7, 2, 7, 15, 7]
target4 = 9
a1 = twoSum(nums1, target1)
a2 = twoSum(nums2, target2)
a3 = twoSum(nums3, target3)
a4 = twoSum(nums4, target4)
print(a1, a2, a3, a4)
```

If more than one solution exists:

```python
def twoSum(nums, target):
    h = {}
    results = [] # list of indices that fulfill the problem
    for i in range(len(nums)):
        h[nums[i]] = i
    for i in range(0, len(nums)):
        y = target - nums[i]
        if y in h and h[y] != i:
            results.append([i, h[y]])
    # return empty list if no valid pair is found
    return results

# test cases
nums1 = [2, 7, 11, 15]
target1 = 9
nums2 = [3, 2, 4]
target2 = 6
nums3 = [3, 3]
target3 = 6
nums4 = [2, 7, 11, 15, 11, 11, 11, 7, 2, 7, 15, 7]
target4 = 9
a1 = twoSum(nums1, target1)
a2 = twoSum(nums2, target2)
a3 = twoSum(nums3, target3)
a4 = twoSum(nums4, target4)
print(a1, a2, a3, a4)
```

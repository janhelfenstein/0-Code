add two numbers
You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

```python
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: Optional[ListNode]
        :type l2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        # create number for list l1, which is in reversed order
        num1 = 0
        num2 = 0
        for i in  range(len(l1)):
            num1 += 10**i * l1[i]
        for i in range(len(l2)):
            num2 += 10**i * l2[i]
        
        # add the two numbers
        result = int(num1 + num2)

        # create reversed list for result
        digits = max(len(l1), len(l2)) + 1
        resultlist = [0] * digits
        for i in range(0, digits):
            resultlist[i] = result % 10 # store the remainder in the list
            result -= resultlist[i]
            result = result / 10
            if result == 0:
                break
        if resultlist[-1] == 0:
            resultlist.pop(-1) # remove last digit in the list since it's 0
        return resultlist

# test cases
self = []
l1 = [2,4,3]
l2 = [5,6,4]
nums = addTwoNumbers(self, l1, l2)
print(nums)

l1 = [0]
l2 = [0]
nums = addTwoNumbers(self, l1, l2)
print(nums)

l1 = [9,9,9,9,9,9,9]
l2 = [9,9,9,9]
nums = addTwoNumbers(self, l1, l2)
print(nums)
```
